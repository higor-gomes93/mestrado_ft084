#### 5) Criação de um Dataset Normalizado
A fim de evitar que atributos com valores em intervalos maiores tenham peso maior que os demais.

```python
x = dataset.values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
n_dataset = pd.DataFrame(x_scaled, columns = columns_names)
n_dataset
```

#### 6) Divisão entre Classe e Atributos
Definir dentro do dataset qual é a variável que será classificada, e quais são as variáveis preditoras.

```python
# Definição das Classes
classes = dataset['drinks']
n_classes = n_dataset['drinks']

# Definição dos atributos
attributes_names = [
    'mcv',
    'alkphos',
    'sgpt',
    'sgot',
    'gammagt'
]

attributes = dataset[attributes_names]
n_attributes = n_dataset[attributes_names]
```

#### 6) Separação dos Dados
Aqui, os dados serão separados em treinamento e teste. O parâmetro *test_size* será definido primeiramente como 0.2 (20%) de acordo com o enunciado da tarefa.

```python
# Divisão da base de dados entre treinamento e teste. Random_state = 0 para sempre obter a mesma divisão da base quando o código for executado.
x_train, x_test, y_train, y_test = train_test_split(attributes, classes, test_size = 0.3, random_state = 0)
n_x_train, n_x_test, n_y_train, n_y_test = train_test_split(n_attributes, n_classes, test_size = 0.3, random_state = 0)
```

#### 7) Árvore de Decisão
Criaçao de uma função que irá encapsular o processo de criação do modelo de árvore de decisão, retornando a classe prevista para cada instância e o erro de classificação do algoritmo.

```python
# Serão utilizados como parâmetros da função os valores de x e y para teste e treinamento
def decision_tree(x_train, x_test, y_train, y_test):
    # Construção do modelo
    tree = DecisionTreeClassifier()
    tree.fit(x_train, y_train)
    
    # Obtenção das previsões
    prediction = tree.predict(x_test)
    
    # Matriz de confusão
    confusion = confusion_matrix(y_test, prediction)
    
    # Taxa acerto
    accuracy = accuracy_score(y_test, prediction)

    # Taxa erro
    error = 1 - accuracy
    
    return prediction, error
```

#### 8) Naïve Bayes
Criaçao de uma função que irá encapsular o processo de criação do modelo de Naïve Bayes, retornando a classe prevista para cada instância e o erro de classificação do algoritmo.

```python
# Serão utilizados como parâmetros da função os valores de x e y para teste e treinamento
def naive_bayes(x_train, x_test, y_train, y_test):
    # Construção do modelo
    naive_bayes = GaussianNB()
    naive_bayes.fit(x_train, y_train)
    
    # Obtenção das previsões
    prediction = naive_bayes.predict(x_test)
    
    # Matriz de confusão
    confusion = confusion_matrix(y_test, prediction)
    
    # Taxa acerto
    accuracy = accuracy_score(y_test, prediction)

    # Taxa erro
    error = 1 - accuracy
    
    return prediction, error
```

#### 9) k-NN
Criaçao de uma função que irá encapsular o processo de criação do modelo de k-NN, retornando a classe prevista para cada instância e o erro de classificação do algoritmo.

```python
# Serão utilizados como parâmetros da função os valores de x e y para teste e treinamento
def k_nn(x_train, x_test, y_train, y_test):
    # Construção do modelo
    knn = KNeighborsClassifier(n_neighbors = 3)
    knn.fit(x_train, y_train)
    
    # Obtenção das previsões
    prediction = knn.predict(x_test)
    
    # Matriz de confusão
    confusion = confusion_matrix(y_test, prediction)
    
    # Taxa acerto
    accuracy = accuracy_score(y_test, prediction)

    # Taxa erro
    error = 1 - accuracy
    
    return prediction, error
```

#### 10) Analisando o Erro
Serão feitos 3 testes:
1. Todos os algoritmos utilizando o dataset inicial
2. Todos os algoritmos utilizando o dataset normalizado
3. Apenas o algoritmo k-NN utilizando o dataset normalizado

```python
# Erro para o caso 1
error_decision_tree_1 = decision_tree(x_train, x_test, y_train, y_test)[1]
error_naive_bayes_1 = naive_bayes(x_train, x_test, y_train, y_test)[1]
error_k_nn_1 = k_nn(x_train, x_test, y_train, y_test)[1]
list_1 = [error_decision_tree_1, error_naive_bayes_1, error_k_nn_1]

# Erro para o caso 2
error_decision_tree_2 = decision_tree(n_x_train, n_x_test, n_y_train, n_y_test)[1]
error_naive_bayes_2 = naive_bayes(n_x_train, n_x_test, n_y_train, n_y_test)[1]
error_k_nn_2 = k_nn(n_x_train, n_x_test, n_y_train, n_y_test)[1]
list_2 = [error_decision_tree_2, error_naive_bayes_2, error_k_nn_2]

# Erro para o caso 3
error_decision_tree_3 = decision_tree(x_train, x_test, y_train, y_test)[1]
error_naive_bayes_3 = naive_bayes(x_train, x_test, y_train, y_test)[1]
error_k_nn_3 = k_nn(n_x_train, n_x_test, n_y_train, n_y_test)[1]
list_3 = [error_decision_tree_3, error_naive_bayes_3, error_k_nn_3]

# Impressão dos resultados
columns_names_error = ['Árvore de Decisão', 'Naïve Bayes', 'k-NN']
rows_names_error = ['Caso 1', 'Caso 2', 'Caso 3']
instances = [list_1, list_2, list_3]
results = pd.DataFrame(instances, index = rows_names_error, columns = columns_names_error)
results
```

#### 11) Ensemble
Criação de uma função que irá encapsular o processo de montagem de um Ensemble com os classificadores treinados (via voto majoritário).

```python
def ensemble(decision_tree, naive_bayes, k_nn, test_array):
    list_of_errors = [(decision_tree[i], naive_bayes[i], k_nn[i]) for i in range(len(test_array))]
    prediction_list = [stats.mode(i)[0][0] for i in list_of_errors]
    return prediction_list
```

```python
x = decision_tree(x_train, x_test, y_train, y_test)[0]
y = naive_bayes(x_train, x_test, y_train, y_test)[0]
z = k_nn(x_train, x_test, y_train, y_test)[0]
test_array = x_test
ensemble(x, y, z, test_array)
```
